```python
from prefect import task, flow, get_run_logger
```

```python
@task(description="A simple task that logs a message.", retries=3, retry_delay_seconds=5)
def log_message() -> int:
    x = 4 + 3
    print(f"result is {x}")
    return x
```

```python
@flow(name="Simple Logging Flow")
def main():
        log_message()

main.serve(
        name="simple-logging-flow",
        host="localhost",
        port=8000,
        cron="0 0 * * *"
)
```
